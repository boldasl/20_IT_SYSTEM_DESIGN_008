```
# numpy, tensorflow, cv2
import numpy as np
from scipy.signal import find_peaks
import tensorflow as tf
import cv2

# keras
from tensorflow import keras
from keras.models import Sequential, save_model, load_model
from keras.layers import Dense, Flatten, Dropout, BatchNormalization
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.callbacks import EarlyStopping
from keras.utils import to_categorical
from keras.preprocessing.image import *

# utility
import matplotlib.pyplot as plt
from timeit import default_timer
from string import ascii_uppercase, ascii_lowercase
from google.colab.patches import cv2_imshow # alternative of cv2.imshow() in jupyter notebook

print("### Tensorflow Version:", tf.__version__)
```

```
# file path config
MAIN_DIR = "./drive/My Drive/Colab Notebooks/HW4_Hangeul_OCR/"

TRAINING_DATA_PATH = "training_data/"
VALIDATION_DATA_PATH = "validation_data/"
SAVED_MODEL_PATH = "saved_model/"

HANGEUL_CLASS_FILE = "hangeul_class.npy"
HANGEUL_TRAINING_IMAGE_FILE = "hangeul_image.npy"
HANGEUL_TRAINING_LABEL_FILE = "hangeul_label.npy"

MNIST_TRAINING_IMAGE_FILE = "mnist_image.npy"
MNIST_TRAINING_LABEL_FILE = "mnist_label.npy"


# model config
INPUT_SIZE = (32, 32, 1)
CLASS_SIZE = 990
MODEL = "CNN1"
RESIZE = (INPUT_SIZE[0], INPUT_SIZE[1])

# training config
TRAIN_FLAG = False
EPOCH = 20
BATCH_SIZE = 64
```

```
#load hangeul class map
mapping = np.load(MAIN_DIR+TRAINING_DATA_PATH+HANGEUL_CLASS_FILE)
print("\"{}\" Load Complete.".format(HANGEUL_CLASS_FILE))

number = np.array([i for i in range(10)])
temp = np.concatenate((number, mapping), axis=0)
mapping = temp

print("-- Class Size: {}".format(mapping.shape))
```

```
# load hangeul, train data, train label
hangeul_train_images = np.load(MAIN_DIR+TRAINING_DATA_PATH+HANGEUL_TRAINING_IMAGE_FILE)
print("\"{}\" Load Complete.".format(HANGEUL_TRAINING_IMAGE_FILE))
hangeul_train_labels = np.load(MAIN_DIR+TRAINING_DATA_PATH+HANGEUL_TRAINING_LABEL_FILE)
print("\"{}\" Load Complete.".format(HANGEUL_TRAINING_LABEL_FILE))

hangeul_train_images = (255 - hangeul_train_images.reshape(-1, INPUT_SIZE[0], INPUT_SIZE[1], INPUT_SIZE[2])) / 255.0
hangeul_train_labels = hangeul_train_labels + 10

if INPUT_SIZE[0] != hangeul_train_images.shape[1] or INPUT_SIZE[1] != hangeul_train_images.shape[2]: print("Hangeul Data Size Mismatch.")

print("-- Train Data Size: {}".format(hangeul_train_images.shape))
print("-- Train Label Size: {}".format(hangeul_train_labels.shape))
```

```
# load mnist(digits) train data, train label
mnist_train_images = np.load(MAIN_DIR+TRAINING_DATA_PATH+MNIST_TRAINING_IMAGE_FILE)
print("\"{}\" Load Complete.".format(MNIST_TRAINING_IMAGE_FILE))
mnist_train_labels = np.load(MAIN_DIR+TRAINING_DATA_PATH+MNIST_TRAINING_LABEL_FILE)
print("\"{}\" Load Complete.".format(MNIST_TRAINING_LABEL_FILE))

mnist_train_images = mnist_train_images.reshape(-1, 32, 32, 1)
#mnist_train_labels = np.argmax(mnist_train_labels, axis=0)

if INPUT_SIZE[0] != mnist_train_images.shape[1] or INPUT_SIZE[1] != mnist_train_images.shape[2]: print("MNIST Data Size Mismatch.")


print("-- Train Data Size: {}".format(mnist_train_images.shape))
print("-- Train Label Size: {}".format(mnist_train_labels.shape))
```

```
train_images = np.concatenate((mnist_train_images, hangeul_train_images))
train_labels = np.concatenate((mnist_train_labels, hangeul_train_labels))
#train_labels = np.where(train_labels < 10, 0, 1)

train_labels = to_categorical(train_labels)

hangeul_train_images = None # free memory
hangeul_train_labels = None
mnist_train_images = None
mnist_train_labels = None

print("-- Class Size: {}".format(mapping.shape))
print("-- Train Data Size: {}".format(train_images.shape))
print("-- Train Label Size: {}".format(train_labels.shape))
```

```
RotationRange = 45
ZoomRange = (1.3, 1.3)

# data augmentation
# training data augmentation
train_images_aug = []
for idx in range(train_images.shape[0]): 
    temp = random_rotation(train_images[idx], rg=RotationRange, row_axis=0, col_axis=1, channel_axis=2)
    temp = random_zoom(temp, zoom_range=ZoomRange, row_axis=0, col_axis=1, channel_axis=2)
    train_images_aug.append(temp)
    if idx % 100 == 0: print("Train Image Augmentation Processing... {0:0.1f}%\r".format(idx/train_images.shape[0]*100), end='')
print("Train Image Augmentation Complete!                         ")

# convert to numpy array
train_images_aug = np.array(train_images_aug)
```

```
_train_images = np.concatenate((train_images, train_images_aug), axis=0)
print(_train_images.shape)
train_images = _train_images
train_images_aug = None # free memory

_train_labels = np.concatenate((train_labels, train_labels), axis=0)
train_labels = _train_labels
print(train_labels.shape)
```

```
# model construction
def network():
    if MODEL is "CNN1":
        model = Sequential()
        model.add(Conv2D(64, (5,5), padding="same", activation="relu", input_shape=INPUT_SIZE))
        model.add(MaxPooling2D(2,2))
        model.add(Conv2D(128, (3,3), padding="same", activation="relu"))
        model.add(MaxPooling2D(2,2))

        model.add(Flatten())
        model.add(Dense(2048, activation="relu"))
        model.add(Dropout(0.4))
        model.add(Dense(1024, activation="relu"))
        model.add(Dense(CLASS_SIZE, activation="softmax"))

    elif MODEL is "VGG11":
        model = Sequential()
        model.add(Conv2D(64, (3,3), activation="relu", padding="same", input_shape=INPUT_SIZE))
        model.add(MaxPooling2D((2,2), strides=(2,2)))

        model.add(Conv2D(128, (3,3), activation="relu", padding="same"))
        model.add(MaxPooling2D((2,2), strides=(2,2)))

        model.add(Conv2D(256, (3,3), activation="relu", padding="same"))
        model.add(Conv2D(256, (3,3), activation="relu", padding="same"))
        model.add(MaxPooling2D((2,2), strides=(2,2)))

        model.add(Conv2D(512, (3,3), activation="relu", padding="same"))
        model.add(Conv2D(512, (3,3), activation="relu", padding="same"))
        model.add(MaxPooling2D((2,2), strides=(2,2)))

        model.add(Conv2D(512, (3,3), activation="relu", padding="same"))
        model.add(Conv2D(512, (3,3), activation="relu", padding="same"))
        model.add(MaxPooling2D((2,2), strides=(2,2), padding="same"))
        
        model.add(Flatten())
        model.add(Dense(4096, activation="relu"))
        model.add(Dense(4096, activation="relu"))
        model.add(Dense(CLASS_SIZE, activation="softmax"))

    elif MODEL is "VGG16":
        model = Sequential()
        model.add(Conv2D(64, (3,3), activation="relu", padding="same", input_shape=INPUT_SIZE))
        model.add(Conv2D(64, (3,3), activation="relu", padding="same"))
        model.add(MaxPooling2D((2,2), strides=(2,2)))

        model.add(Conv2D(128, (3,3), activation="relu", padding="same"))
        model.add(Conv2D(128, (3,3), activation="relu", padding="same"))
        model.add(MaxPooling2D((2,2), strides=(2,2)))

        model.add(Conv2D(256, (3,3), activation="relu", padding="same"))
        model.add(Conv2D(256, (3,3), activation="relu", padding="same"))
        model.add(Conv2D(256, (3,3), activation="relu", padding="same"))
        model.add(MaxPooling2D((2,2), strides=(2,2)))

        model.add(Conv2D(512, (3,3), activation="relu", padding="same"))
        model.add(Conv2D(512, (3,3), activation="relu", padding="same"))
        model.add(Conv2D(512, (3,3), activation="relu", padding="same"))
        model.add(MaxPooling2D((2,2), strides=(2,2)))

        model.add(Conv2D(512, (3,3), activation="relu", padding="same"))
        model.add(Conv2D(512, (3,3), activation="relu", padding="same"))
        model.add(Conv2D(512, (3,3), activation="relu", padding="same"))
        model.add(MaxPooling2D((2,2), strides=(2,2), padding="same"))

        model.add(Flatten())
        model.add(Dense(4096, activation="relu"))
        model.add(Dense(4096, activation="relu"))
        model.add(Dense(CLASS_SIZE, activation="softmax"))

    return model

print("{} Model Contructed.".format(MODEL))
```

```
if TRAIN_FLAG:
    model = network()
    model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
    print(model.summary())
    print("Training Started.")
    hist = model.fit(train_images, train_labels, epochs=EPOCH, batch_size=BATCH_SIZE)
    MODEL_FILE = "{}_CLASS-{}_EPOCH-{}_BATCH-{}.json".format(MODEL, CLASS_SIZE, EPOCH, BATCH_SIZE)
    save_model(model, MAIN_DIR+SAVED_MODEL_PATH+MODEL_FILE)
else: 
    LOAD_MODEL_FILE = "CNN1_CLASS-990_EPOCH-20_BATCH-64.json"
    model = load_model(MAIN_DIR+SAVED_MODEL_PATH+LOAD_MODEL_FILE)
    print(model.summary())



if TRAIN_FLAG:
    log_loss = hist.history["loss"]
    plt.plot(range(len(log_loss)), log_loss)
    plt.xlabel("epoch")
    plt.ylabel("loss")
    plt.show()
```

```
# image file import & find boxes
raw_image = cv2.imread(MAIN_DIR+VALIDATION_DATA_PATH+"한글손글씨_10.jpg", cv2.IMREAD_COLOR)

gray_image = cv2.cvtColor(raw_image, cv2.COLOR_BGR2GRAY)
print("↓↓↓ Gray-scale Image ↓↓↓"); cv2_imshow(gray_image) 

blurred_image = cv2.blur(gray_image, (3, 3)) # blur
print("\n\n↓↓↓ Blurred Image ↓↓↓"); cv2_imshow(blurred_image)

inverted_image = 255 - blurred_image # inverse image
print("\n\n↓↓↓ Inverted Image ↓↓↓"); cv2_imshow(inverted_image)

_, threshold_image = cv2.threshold(inverted_image.copy(), 120, 255, cv2.THRESH_TOZERO)
print("\n\n↓↓↓ Threshold Image ↓↓↓"); cv2_imshow(threshold_image)


closed_kernel = np.ones((15,15))
closed_image = cv2.morphologyEx(threshold_image.copy(), cv2.MORPH_CLOSE, closed_kernel)
print("\n\n↓↓↓ Closed Image ↓↓↓"); cv2_imshow(closed_image)

'''
eroded_kernel = np.ones((3,3))
eroded_image = cv2.morphologyEx(threshold_image.copy(), cv2.MORPH_ERODE, eroded_kernel)
print("\n\n↓↓↓ Eroded Image ↓↓↓"); cv2_imshow(eroded_image)
'''

# finding boxes
contours, _ = cv2.findContours(closed_image.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
boxes = [cv2.boundingRect(each) for each in contours]
boxes = np.array(boxes)


# filtering boxes
size = boxes[:, 2:] # contour를 따라 찾은 box에서 가로 세로 길이만 추출
area = np.multiply(size[:, 0], size[:, 1]) # rectangular의 넓이를 계산
mean_area = np.mean(area) # 모든 rectangular의 넓이를 평균냄

# 평균 넓이의 0.2배 이상의 box만 남김
# noise에 의해 발생하는 작은 box들을 제거
filtered_box = []
for i in range(boxes.shape[0]):
    if boxes[i][2]*boxes[i][3] > 0.3 * mean_area:
        filtered_box.append([boxes[i][j] for j in range(4)])
boxes = np.array(filtered_box)


boxed_image = raw_image
green = (0, 255, 0)
for box in boxes:
    cv2.rectangle(boxed_image, (box[0], box[1]), (box[0]+box[2], box[1]+box[3]), green, 2)
print("\n\n↓↓↓ Boxed Image ↓↓↓"); cv2_imshow(boxed_image)
```

```
# 위에서 아래로, 좌에서 우로 정렬
image_size = gray_image.shape
syllable_length = boxes.shape[0]

print("image_size =", image_size)
print("syllable_length =", syllable_length)

#sorted_boxes = boxes[np.argsort(boxes[:, 1])] # 세로축 기준으로 정렬
#print(sorted_boxes)

# 세로축을 훑으며 박스가 포함된 개수 세기
box_count = []
for idx_y in range(image_size[0]):
    cnt = 0
    for _, coord, _, length in boxes:
        if (idx_y > coord) and (idx_y < coord+length): cnt += 1
    box_count.append(cnt)

plt.plot(range(len(box_count)), box_count)
plt.show()

# peaks 찾기
peaks, _ = find_peaks(box_count)

# 정렬하기
sorted_boxes = [[] for i in range(len(peaks))]

for idx, peak in enumerate(peaks):
    for box in boxes:
        if (peak > box[1]) and (peak < box[1]+box[3]):
            sorted_boxes[idx].append(box)
    sorted_boxes[idx] = np.array(sorted_boxes[idx])
sorted_boxes = np.array(sorted_boxes)

row_count = []
for idx in range(len(peaks)):
    row_count.append(len(sorted_boxes[idx]))

for idx in range(1, len(row_count)):
    row_count[idx] += row_count[idx-1]
print(row_count)

for idx in range(sorted_boxes.shape[0]):
    sorted_boxes[idx] = sorted_boxes[idx][np.argsort(sorted_boxes[idx][:, 0])]    

arranged_boxes = []
for idx in range(sorted_boxes.shape[0]):
    arranged_boxes.extend(sorted_boxes[idx])
```

```

# resizing with ratio & padding

contents = []
for idx in range(syllable_length):
    raw_box = gray_image[arranged_boxes[idx][1]:arranged_boxes[idx][1]+arranged_boxes[idx][3], 
                         arranged_boxes[idx][0]:arranged_boxes[idx][0]+arranged_boxes[idx][2]]
    raw_ratio = raw_box.shape

    # 길이가 긴쪽을 기준으로 하여 resizing
    shrink_size = 26
    if raw_ratio[0] > raw_ratio[1]: size = (raw_ratio[1]*shrink_size//raw_ratio[0], shrink_size)
    else: size = (shrink_size, raw_ratio[0]*shrink_size//raw_ratio[1])
    
    resized_box = cv2.resize(raw_box, dsize=size, interpolation=cv2.INTER_AREA)
    resized_ratio = resized_box.shape

    pad_top = (RESIZE[0] - resized_ratio[0]) // 2
    pad_bottom = RESIZE[0] - resized_ratio[0] - pad_top
    pad_left = (RESIZE[1] - resized_ratio[1]) // 2
    pad_right = RESIZE[1] - resized_ratio[1] - pad_left

    padded_box = cv2.copyMakeBorder(resized_box, 
                                    pad_top, pad_bottom, 
                                    pad_left, pad_right, 
                                    borderType=cv2.BORDER_CONSTANT, 
                                    value=[255])
    
    inverted_box = (255 - padded_box)

    _, threshold_box = cv2.threshold(inverted_box, 100, 255, cv2.THRESH_TOZERO)
    
    

    

    cv2_imshow(threshold_box)
    contents.append(threshold_box.reshape(INPUT_SIZE) / 255.0)
contents = np.array(contents)
print("-- Contents Size: {}".format(contents.shape))
```

```
predicted = model.predict(contents)
sentences = []

for idx in range(predicted.shape[0]):
    sentences.append(mapping[np.argmax(predicted[idx])])
    print(mapping[np.argmax(predicted[idx])], end='')
    if row_count.count(idx+1): 
        sentences.append('\n')
        print()
```
